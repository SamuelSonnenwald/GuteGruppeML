```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.linear_model import LogisticRegression

from sklearn.linear_model import Ridge
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.feature_selection import RFECV, SelectKBest
from sklearn.preprocessing import StandardScaler


maindata = pd.read_csv("https://archive.ics.uci.edu/static/public/2/data.csv")
#maindata.head()

maindata_processed = maindata.copy()

data_target = maindata["income"]

x = data_target
y = maindata_processed

maindata_processed = maindata.drop(columns="income")

label_encoder = preprocessing.LabelEncoder()
data_target = label_encoder.fit_transform(data_target)

encoder = preprocessing.OneHotEncoder()
encoded = pd.DataFrame(encoder.fit_transform(maindata_processed[["workclass","education","marital-status","occupation","relationship","race","sex","native-country"]]).toarray(), columns=encoder.get_feature_names_out(["workclass","education","marital-status","occupation","relationship","race","sex","native-country"]))

maindata_processed = maindata_processed.drop(columns=["workclass","education","marital-status","occupation","relationship","race","sex","native-country"])
maindata_processed = maindata_processed.join(encoded)

data_train, data_test, target_train, target_test = train_test_split(
    maindata_processed, data_target,test_size=0.2, random_state=42, stratify=data_target)

display(data_test)
display(data_train)
logit = LogisticRegression(max_iter = 10000)

print(logit.fit(x,y))

print(logit.score(x,y))
```
